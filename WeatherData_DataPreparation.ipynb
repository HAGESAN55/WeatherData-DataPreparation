# Import Module

```python
import numpy as np
import pandas as pd
```

# Import data cuaca
data cuaca tahun 2016 diinput kedalam dataset dengan dirubah setiap kolom menjadi bahasa Indonesia

```python
df = pd.read_csv("2016-weather.csv")
kolom = ['Tanggal', 'Bulan', 'Rata-rata Temperatur', 'Total Hujan (cm)', 'Total Salju (cm)', 'Tinggi Salju (cm)']
df = df.set_axis(kolom, axis=1, inplace=False)
df
```

# Missing Value Checking
Melihat data kosong pada dataset, terdapat beberapa data yang kosong pada beberapa kolom pada dataset

```python
df.apply(lambda x: sum(x.isnull()), axis = 0)
```

# Data Preparation
Menyiapkan data agar data dapat diolah kedalam model, persiapan yang dilakukan antara lain:
- Pengisian data kosong dengan parameter kelompok adalah bulan

```python
# Mengcopy Dataset kedalam variabel data

data = df.copy().drop('Tanggal', axis=1)
data
```

```python
# Mengkelompokan data berdasarkan bulannya, dan melihat rata rata nilai kolomnya berdasarkan bulan

df.groupby('Bulan').mean()
```

```python
# Mengisi data kosong pada setiap kolom, dengan isian data yang kosong berdasarkan nilai rata rata setiap bulannya

for x in data['Bulan'].unique():
  for y in data.columns:
    g = df[df['Bulan']==x]
    g[y].fillna(round(g[y].mean(), 2), inplace=True)
    df[df['Bulan']==x] = g

df
```

```python
# Menampilkan kembali data yang kosong

df.apply(lambda x: sum(x.isnull()), axis=0)
```

```python
# Menampilkan dataset berdasarkan bulan dan rata rata nilai setiap kelompoknya

df.groupby('Bulan').mean()
```

```python
# Menampilkan dataset berdasarkan nilai temperatur yang lebih kecil dari 4 derajat

df[df['Rata-rata Temperatur']<4]
```

```python
# Menampilakan dataset yang berdasarkan bulan ke 3

df[df["Bulan"]==3]
```
